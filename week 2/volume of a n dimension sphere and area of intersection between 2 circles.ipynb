```python
import numpy as np
import matplotlib.pyplot as plt

#volume of an N dimension sphere
def get_volume(N):

    #uses 1000000 points in monte carlo simulation
    points=np.random.uniform(-1,1,size=(1000000,N))
    distances=np.sum(points**2,axis=1)
    inside_mask=distances<=1
    count=np.sum(inside_mask)

    #using N dimension arrays
    return (2**N)*count/1000000

dimensions=np.arange(2,11,dtype='int')
volume=np.vectorize(get_volume)(dimensions)

array=np.column_stack((dimensions,volume))
print(array)
```

```python
import numpy as np

def area_of_intersection(x1,y1,r1,x2,y2,r2,N=100000):

#pre condition (r1-r2)**2<(x1-x2)**2+(y1-y2)**2<(r1+r2)**2
#making a quadrilateral that must contain both the circles in it
#area=(y_high-y_low)*(x_high-x_low)

  x_low=min(x1,x2)-max(r1,r2)
  x_high=max(x1,x2)+max(r1,r2)
  y_low=min(y1,y2)-max(r1,r2)
  y_high=max(y1,y2)+max(r1,r2)

  #generating points in the quadrilateral
  x=np.random.uniform(x_low,x_high,N)
  y=np.random.uniform(y_low,y_high,N)

  #masking the points present within the area of intersection
  logical_mask1=(x-x1)**2+(y-y1)**2<=r1**2
  logical_mask2=(x-x2)**2+(y-y2)**2<=r2**2

  logical_mask=logical_mask1*logical_mask2

  count=np.sum(logical_mask)

  area=(y_high-y_low)*(x_high-x_low)*count/N

  return area

# x1,y1,r1,x2,y2,r2=variables
print(area_of_intersection(0,0,1,0,0,3))
```
