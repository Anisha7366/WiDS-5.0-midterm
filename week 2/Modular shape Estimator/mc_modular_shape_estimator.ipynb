```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
from numpy import sqrt,pi
from mc_modular_shape_estimator import run_simulation,is_in_circle,is_in_gaussian,is_in_parabola

calculated_value=(sqrt(pi)/2)*erf(2)
```

```python
#plotting convergence error of the 3 curves using numbers from 2 to 10000 as input

fig, axes=plt.subplots(1,3,figsize=(15,6))

inputs=np.arange(2,10000)

ax=axes[0]

outputs=np.vectorize(run_simulation,excluded=[0,1])(is_in_circle, [-1,1,-1,1], inputs)

x=inputs
y=100*abs(outputs-np.pi)/np.pi

ax.plot(x,y,color='black',label='Error percentage')

ax.set_xlabel('NUMBERS')
ax.set_ylabel('CALCULATED ERROR PERCENTAGE')
ax.set_title('Convergence error plot')

ax.axhline(0,color='red',ls='--',label="0 ERROR PERCENTAGE")

ax.legend()

ax=axes[1]

outputs=np.vectorize(run_simulation,excluded=[0,1])(is_in_parabola,[0,1,0,1], inputs)

x=inputs
y=100*abs(outputs-(1/3))*3

ax.plot(x,y,color='black',label='Error Percentage')

ax.set_xlabel('NUMBERS')
ax.set_ylabel('CALCULATED ERROR PERCENTAGE')
ax.set_title('Convergence error plot')

ax.axhline(0,color='red',ls='--',label="0 ERROR PERCENTAGE")

ax.legend()

ax=axes[2]

outputs=np.vectorize(run_simulation,excluded=[0,1])(is_in_gaussian,[0,2,0,1], inputs)

x=inputs
y=100*abs(outputs-calculated_value)/calculated_value

ax.plot(x,y,color='black',label='Error Percentage')

ax.set_xlabel('NUMBERS')
ax.set_ylabel('CALCULATED ERROR PERCENTAGE')
ax.set_title('Convergence error plot')

ax.axhline(0,color='red',ls='--',label="0 ERROR PERCENTAGE")

ax.legend()

plt.tight_layout()
plt.show()
```
