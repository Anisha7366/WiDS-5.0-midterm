```python
import numpy as np

#TASK A: AREA UNDER SINE CURVE FROM -1 TO 1
# REIMANN SUM
# ANALYTICAL VALUE=2
x=np.linspace(0,np.pi,1000,endpoint=True)
y=np.sin(x)

y=y*(x[1]-x[0])

area=np.sum(y)

abs_error=abs(area-2)
print(abs_error)
```

```python
#TASK A: AREA UNDER SINE CURVE FROM -1 TO 1
# MONTE CARLO

x=np.random.uniform(0,np.pi,100000)
y=np.random.uniform(0,1,100000)

inside_mask=y<np.sin(x)
count=np.sum(inside_mask)

area=np.pi*count/100000

abs_error=abs(area-2)
print(abs_error)
```

Reimann sum is much more accurate(task A)

```python
import numpy as np

#TASK B: VOLUME OF A 10 DIMENSIONAL UNIT HYPERSPHERE
# REIMANN SUM
# ANALYTICAL VALUE=2.55

#GENERATING GRID POINTS AT -0.75,-0.25,0.25,0.75 FOR ACCURACY, MAKING A GRID WITH ALL POSSIBLE COMBINATIONS AND FINDING SUM
grid_points=np.linspace(-0.75,0.75,4)

grid=np.meshgrid(*([grid_points]*10),indexing="ij")

points=np.stack(grid,axis=-1).reshape(-1,10)

point_squares=np.sum(points**2,axis=1)

mask=point_squares<=1

#spacing is= 0.5 so volume of each 10D sun unit in the grid=0.5**10

vol=np.sum(mask)*((0.5))**10
abs_error=abs(vol-2.55)
print(abs_error)
```

```python
#TASK B: VOLUME OF A 10 DIMENSIONAL UNIT HYPERSPHERE
# MONTE CARLO

#generating a 10D box with vol=2**10 and 10,00,000 points
points=np.random.uniform(-1,1,size=(1000000,10))
distances=np.sum(points**2,axis=1)
#points inside sphere
inside_mask=distances<=1
count=np.sum(inside_mask)

volume=(2**10)*count/1000000
abs_error=abs(volume-2.55)
print(abs_error)
```

Monte Carlo is so much more accurate compared to the reimann sum

```python
#TASK C: 6D GAUSSIAN
# ANALYTICAL VALUE/TRUE VALUE

from scipy.special import erf
import numpy as np
true_val = (np.sqrt(np.pi) * erf(1)) ** 6
true_val
```

```python
#TASK C: 6D GAUSSIAN
# TAYLOR SERIES APPROXIMATION UPTO 8 TERMS
# ANALYTICAL VALUE=2.55

import numpy as np
from scipy.integrate import quad

def f(x):
    return 1-x**2+x**4/2-x**6/6+x**8/24-x**10/120+x**12/720-x**14/5040

result, error = quad(f, -1, 1)

area=result**6
abs_error=abs(area-true_val)
abs_error
```

```python
#TASK C: 6D GAUSSIAN
# RIEMANN SUM
# ANALYTICAL VALUE=2.55

#making a grid with 10 points over 6 dimensions each,volume accordingly
import numpy as np

axes=np.linspace(-1,1,10)

grid=np.meshgrid(*([axes]*6),indexing="ij")

points=np.stack(grid,axis=-1).reshape(-1,6)

point_squares=np.sum(points**2,axis=1)
e_power=np.e**(-point_squares)

count=np.sum(e_power)

area=((axes[1]-axes[0])**6)*count
abs_error=abs(area-true_val)
abs_error
```

```python
#TASK C: 6D GAUSSIAN
# MONTE CARLO
# ANALYTICAL VALUE=2.55

#instead of a defined boundary this just sums the value of gaussian over the generated points and averages that and scales accordingly
import numpy as np

points=np.random.uniform(-1,1,size=(1000000,6))
distances=np.sum(points**2,axis=1)
e_power=np.e**(-distances)

sum=np.sum(e_power)

volume=(2**6)*sum/1000000
abs_error=abs(volume-true_val)
abs_error
```

```python
#The taylor series approximation was very close, it does get worse if the limits are increased as shown accordingly:

from scipy.special import erf
import numpy as np
true_val = (np.sqrt(np.pi) * erf(1)) ** 6

from scipy.integrate import quad

def f(x):
    return 1-x**2+x**4/2-x**6/6+x**8/24-x**10/120+x**12/720-x**14/5040

result, error = quad(f, -5, 5)

area=result**6
abs_error=abs(area-true_val)
abs_error
```

hence taylor's series approximation was pretty good but on increasing the limits the approximation get's bad, the reimann sum was not very accurate compared to monte carlo
