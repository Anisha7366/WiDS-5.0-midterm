```python
import numpy as np
from mc_e import get_e_method1,get_e_method2

#generating numbers from 2 to 100 and running the code on all of them, prints mean value of e calculated and mean percentage error

numbers=np.arange(2,100)

values_of_e1=np.vectorize(get_e_method1)(numbers,100000)
print("estimated value of e:",np.mean(values_of_e1))

percentage_error=100*abs(np.e-values_of_e1)/np.e
print("percentage error:",np.mean(percentage_error),"%")
```

```python
#running the code on number of choice

number=int(input("Enter number: "))
print("estimated value of e:",get_e_method1(number,100000))
```
