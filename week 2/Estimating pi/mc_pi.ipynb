```python
import numpy as np
import matplotlib.pyplot as plt
from mc_pi import get_pi
```

```python
#generating powers of 10 from 10 to 10 million and calculating pi using them as inputs

fig, axes=plt.subplots(1,2,figsize=(13,6))

x=np.arange(1,8,1,dtype='int')
inputs=10**x
y1=np.vectorize(get_pi)(inputs)
y2=abs(y1-np.pi)*100/np.pi

ax=axes[0]

ax.plot(x,y1,color='black',marker='.',ms=12,label='Estimated value')

ax.set_xlabel('N (Log10 Scale)')
ax.set_ylabel('Estimated pi')
ax.set_title('Estimating pi')

ax.axhline(np.pi,color='red',ls='--',label="True value of pi")

ticks=ax.get_yticks()
ticks=np.append(ticks,np.pi)
ticks=np.sort(ticks)
ax.set_yticks(ticks)

ax.legend()

ax=axes[1]

ax.plot(x,y2,color='black',marker='.',ms=12,label='Error Percentage')

ax.set_xlabel('N (Log10 Scale)')
ax.set_ylabel('Percentage Error')
ax.set_title('Estimated Error')

ax.legend()

plt.tight_layout()
plt.show()
```

why did vectorization in place of loops make the code run faster? avoids runtime typechecking and performs the same operation on many elements simultaneously instead of one by one(as in the case of loops).
